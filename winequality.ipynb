```python
from sklearn.model_selection import train_test_split

train_set,test_set=train_test_split(df,train_size=0.99,shuffle=True,random_state=100)
```

```python
X_train=train_set[["alcohol","density","pH","chlorides",
                   "residual.sugar","volatile.acidity","fixed.acidity"]]
Y_train=test_set["quality"]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score
from sklearn.metrics import accuracy_score
```

```python
X_train=train_set[["alcohol","density","pH","chlorides",
                   "residual.sugar","volatile.acidity","fixed.acidity"]]
Y_train=train_set["quality"]
X_test=test_set[["alcohol","density","pH","chlorides",
                   "residual.sugar","volatile.acidity","fixed.acidity"]]
Y_test = test_set['quality']
```

```python
clf=LinearRegression()
clf.fit(X_train,Y_train)
predicted_Y=clf.predict(X_test)
predicted=np.round(predicted_Y)
accuracy_score(predicted,Y_test)
```

```python
clf= LogisticRegression()
clf.fit(X_train,Y_train)
predicted_Y=clf.predict(X_test)
predicted=np.round(predicted_Y)
accuracy_score(predicted,Y_test)
```

```python
clf=RandomForestRegressor()
clf.fit(X_train,Y_train)
predicted_Y=clf.predict(X_test)
predicted=np.round(predicted_Y)
accuracy_score(predicted,Y_test)
```

```python
clf=SVC(kernel="linear")
clf.fit(X_train,Y_train)
predicted_Y=clf.predict(X_test)
predicted=np.round(predicted_Y)
accuracy_score(predicted,Y_test)
```

```python
clf=RandomForestRegressor()
clf.fit(X_train,Y_train)
predicted_Y=clf.predict(X_test)
predicted=np.round(predicted_Y)
accuracy_score(predicted,Y_test)
```

```python
clf=DecisionTreeRegressor()
clf.fit(X_train,Y_train)
predicted_Y=clf.predict(X_test)
predicted=np.round(predicted_Y)
accuracy_score(predicted,Y_test)
```

Here we are training and testing on the same data, the goal is to estimate likely performance of a model on out-of-sample data, but, maximizing training accuracy rewards overly complex models that won't necessarily generalize (Unnecessarily complex models overfit the training data).
So when we evaluate the model we trained we get high scores, this just mean how well our model learnt from our training data.
Moreover to make experimental results more reliable we should clean our data i.e removing null/empty values and clearing duplicates
